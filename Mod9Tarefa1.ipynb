### 2. Peso médio dos bebes por sexo que nasceram no dia do seu aniversário por faixas de escolaridade mae
Ex: Você, aluna(o), nasceu no dia 10/01, então você precisa filtrar o conjunto de dados nessa data e calcular o peso médio dos bebês de cada sexo por faixa de escolaridade da mãe.

```python
sinasc = sinasc_raw[[
    'IDADEMAE', 'SEXO', 'APGAR1', 'APGAR5', 'PESO', 'CONSULTAS', 'DTNASC',
    'GESTACAO', 'GRAVIDEZ', 'ESCMAE', 'IDADEPAI'
]].copy()
sinasc.head(5)
```

```python
#meu aniversário é 02jan
fem = sinasc[(sinasc['DTNASC'] == '2019-01-02') & (sinasc['SEXO'] == 'Feminino')]
peso_medio_fem = fem.groupby('ESCMAE')['PESO'].mean()
print(peso_medio_fem)

```

```python
#meu aniversário é 02jan
masc = sinasc[(sinasc['DTNASC'] == '2019-01-02') & (sinasc['SEXO'] == 'Masculino')]
peso_medio_masc = masc.groupby('ESCMAE')['PESO'].mean()
print(peso_medio_masc)
```

```python


peso_medio = sinasc[sinasc['DTNASC'] == '2019-01-02'].groupby(['ESCMAE', 'SEXO'])['PESO'].mean().unstack()

peso_medio.plot(kind='bar', figsize=[15, 5])

plt.title('Peso Médio dos Bebês nascidos em 2019-01-02, por Escolaridade da Mãe e Sexo')
plt.xlabel('Escolaridade da Mãe')
plt.ylabel('Peso Médio (g)')
plt.legend(title='Sexo')
plt.show()
```

```python
sinasc_2 = sinasc_raw[['munResNome', 'IDADEMAE', 'IDADEPAI']]
sinasc_2.head()
```

```python
sinasc_castanheiras = sinasc_2[sinasc_2['munResNome'] == 'Castanheiras']

idade_mae_media = sinasc_castanheiras['IDADEMAE'].mean()
idade_mae_maxima = sinasc_castanheiras['IDADEMAE'].max()
idade_mae_minima = sinasc_castanheiras['IDADEMAE'].min()

# Calcular as estatísticas dos pais
idade_pai_media = sinasc_castanheiras['IDADEPAI'].mean()
idade_pai_maxima = sinasc_castanheiras['IDADEPAI'].max()
idade_pai_minima = sinasc_castanheiras['IDADEPAI'].min()

# Criar um DataFrame com os resultados
estatisticas = pd.DataFrame({
    'Estatística': ['Média', 'Máxima', 'Mínima'],
    'Idade Mães': [idade_mae_media, idade_mae_maxima, idade_mae_minima],
    'Idade Pais': [idade_pai_media, idade_pai_maxima, idade_pai_minima]
})

print(estatisticas)
```

### 4. Qual o municipio que nasceu mais bebe no mês de março?
    - qual a quantidade de filhos vivos media, maxima, minima nesse municipio?
    - qual a idade media, maxima, minima dos pais nesse municipio?



```python
sinasc_raw['DTNASC'] = pd.to_datetime(sinasc_raw['DTNASC'])
sinasc_raw['MES'] = sinasc_raw['DTNASC'].dt.month

marco_nascimentos = sinasc_raw[sinasc_raw['MES'] == 3]
print(marco_nascimentos)
```

```python
municipios_nascimentos = marco_nascimentos.groupby('munResNome').size().sort_values()
print(municipios_nascimentos)
```

```python
#Porto Velho é o Município que nasceu mais bebês.
marco_nascimentos = sinasc_raw[sinasc_raw['MES'] == 3]
pv = marco_nascimentos[marco_nascimentos['munResNome'] == 'Porto Velho']['QTDFILVIVO'].aggregate(['min', 'mean', 'max'])
print(pv)
```

```python
#A quantidade máxima, média e mínima de filhos vivos das mães que tiveram bebês no mês de março em Porto Velho é 9, 1.29 e 0, respectivamente. 
```

```python
marco_nascimentos = sinasc_raw[sinasc_raw['MES'] == 3]
pvP = marco_nascimentos[marco_nascimentos['munResNome'] == 'Porto Velho']['IDADEPAI'].aggregate(['min', 'mean', 'max'])
print(pvP)
```
